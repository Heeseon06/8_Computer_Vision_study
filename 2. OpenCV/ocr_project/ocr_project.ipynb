# **1.EasyOCR**
* EasyOCR은 오픈소스 OCR(Optical Character Recognition) 라이브러리로, 딥러닝을 기반으로 텍스트를 이미지에서 추출하는 도구
* 다양한 언어 및 문자체 지원
* 이미지에서의 텍스트 추출 및 인식
* 최신 딥러닝 기술을 활용하여 높은 정확도의 텍스트 인식을 제공

```python
import cv2
import easyocr
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# 이미지 파일 경로 지정
image_path ='/content/drive/MyDrive/kdt/ocr팀과제/dd.jpg'

# 이미지를 읽기
image = Image.open(image_path)

# 이미지를 numpy 배열로 변환
img_array = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

# 이미지 표시
plt.figure(figsize=(10, 10))                            # 10x10 인치 크기의 새로운 그래프를 생성
plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))  # 이미지를 BGR에서 RGB로 다시 변환한 후 그래프에 표시. Matplotlib는 RGB 형식을 사용하기 때문
plt.axis('off')                                         # 그래프의 축을 숨김
plt.show()

# OCR 수행
reader = easyocr.Reader(['ko', 'en'])
result = reader.readtext(img_array)

# OCR 결과 출력 및 이미지에 경계 상자 그리기
for (bbox, text, prob) in result:                             # OCR 결과의 경계 상자 (bbox), 텍스트 (text), 신뢰도 (prob)를 포함
    print(f"Detected text: {text} with confidence: {prob}")   # 인식된 텍스트와 그 신뢰도를 출력
    top_left = tuple(map(int, bbox[0]))                       # 경계 상자의 왼쪽 상단 좌표를 정수형 튜플로 변환
    bottom_right = tuple(map(int, bbox[2]))                   # 경계 상자의 오른쪽 하단 좌표를 정수형 튜플로 변환
    cv2.rectangle(img_array, top_left, bottom_right, (0, 255, 0), 2) # 이미지에 녹색((0, 255, 0)) 경계 상자를 그립니다. 선 두께는 2픽셀

# OCR 결과 이미지 표시
plt.figure(figsize=(10, 10))
plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
!wget https://noto-website-2.storage.googleapis.com/pkgs/NotoSansCJKjp-hinted.zip
!unzip NotoSansCJKjp-hinted.zip -d NotoSans
# PaddleOCR의 draw_ocr 함수는 텍스트를 이미지에 그려주는 기능을 제공, 함수는 기본적으로 폰트를 사용하여 인식된 텍스트를 이미지에 그리기 때문에 폰트 파일이 필요
# 폰트를다운받아줌
```

```python
from paddleocr import PaddleOCR, draw_ocr
import matplotlib.pyplot as plt
from PIL import Image

# PaddleOCR 모델 초기화
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # use_angle_cls=True는 텍스트의 각도를 자동으로 조정하도록 설정

# 이미지 파일 경로
image_path = '/content/drive/MyDrive/kdt/ocr팀과제/dd.jpg'

# OCR 수행
result = ocr.ocr(image_path, cls=True)

# 결과 출력
for line in result:
    print(line)

# 결과 시각화
image = Image.open(image_path).convert('RGB')
boxes = [elements[0] for elements in line for line in result]  # 각 텍스트의 경계 상자 좌표를 추출하여 boxes 리스트에 저장
txts = [elements[1][0] for elements in line for line in result] # 인식된 텍스트를 추출하여 txts 리스트에 저장
scores = [elements[1][1] for elements in line for line in result] # 각 텍스트의 신뢰도를 추출하여 scores 리스트에 저장

im_show = draw_ocr(image, boxes, txts, scores, font_path='NotoSans/NotoSansCJKjp-Medium.otf') # draw_ocr 함수를 사용하여 인식된 텍스트와 경계 상자를 이미지에 그림
im_show = Image.fromarray(im_show)

# 이미지 표시
plt.imshow(im_show)
plt.axis('off')
plt.show()
```

# **3.CLOVA OCR**
* CLOVA OCR는 네이버의 인공지능 플랫폼 CLOVA에서 제공하는 OCR(Optical Character Recognition) 서비스입니다.

```python
import requests
import json
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# 네이버 API 설정
api_url = 'https://km4wx7kd2v.apigw.ntruss.com/custom/v1/33023/2cfe71fb603d2cfe51f82be27727ce237e5e92a9199c41c046404d9c3a364e57/general'
secret_key = 'cHZiR09HSlFnWWJmenhEQnZ3S255c2RBb3ZGdFREbEs='

# webp 파일을 열기
image = Image.open('/content/drive/MyDrive/kdt/ocr팀과제/dd.webp')

# jpg로 저장
image.save('/content/drive/MyDrive/kdt/ocr팀과제/dd.jpg', 'JPEG')

# 변환된 jpg 파일의 경로를 image_path에 할당
image_path = '/content/drive/MyDrive/kdt/ocr팀과제/dd.jpg'

# 이미지 파일 읽기
with open(image_path, 'rb') as image_file:
    file_data = image_file.read()

    request_json = {
        'images': [
            {
                'format': 'jpg',
                'name': 'demo'
            }
        ],
        'requestId': 'string',
        'version': 'V2',
        'timestamp': 0
    }                               # 네이버 OCR API에 보낼 요청 본문을 JSON 형식으로 설정, 여기에는 이미지 형식, 이름, 요청 ID, 버전, 타임스탬프 등이 포함

    payload = {'message': json.dumps(request_json).encode('UTF-8')}   # 요청 본문을 JSON 형식으로 인코딩하여 payload 변수에 저장

    files = {
        'file': file_data   # 이미지 파일 데이터를 포함하는 딕셔너리를 생성하여 files 변수에 저장
    }

    headers = {
        'X-OCR-SECRET': secret_key
    }

    # API 요청
    response = requests.post(api_url, headers=headers, files=files, data=payload)   # 설정된 URL, 헤더, 파일, 페이로드를 포함하여 POST 요청을 보내고 응답을 response 변수에 저장
    result = response.json()

# 결과 출력
print(json.dumps(result, indent=4, ensure_ascii=False))

# 결과 시각화 및 텍스트 표시
image = Image.open(image_path)
draw = ImageDraw.Draw(image)

# 폰트 설정 (폰트 파일의 경로를 지정해야 합니다)
# font = ImageFont.truetype("/path/to/font.ttf", 20)  # 폰트 파일 경로와 크기 지정
# 폰트 파일이 없다면 기본 폰트 사용
font = ImageFont.load_default()

# OCR 결과에서 텍스트와 바운딩 박스 추출
if 'images' in result and len(result['images']) > 0:   # OCR 결과에 이미지가 포함되어 있는지 확인
    for item in result['images'][0].get('fields', []):
        text = item['inferText']                      # 인식된 텍스트를 text 변수에 저장
        bbox = item['boundingPoly']['vertices']       # 텍스트의 바운딩 박스 좌표를 bbox 변수에 저장

        # 바운딩 박스 그리기
        draw.polygon([
            (bbox[0]['x'], bbox[0]['y']),
            (bbox[1]['x'], bbox[1]['y']),
            (bbox[2]['x'], bbox[2]['y']),
            (bbox[3]['x'], bbox[3]['y'])
        ], outline="red")

        # 텍스트 그리기
        draw.text((bbox[0]['x'], bbox[0]['y'] - 20), text, font=font, fill="red")  # 텍스트를 바운딩 박스의 위쪽에 그림

plt.figure(figsize=(15,15))
plt.imshow(image)
plt.axis('off')
plt.show()
```
